```python
data_path = '../data'
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


plt.rcParams.update({'font.size': 14})
```

```python
def load_data(path_dict, fname):
    # Placeholder for data loading logic
    # This function should read the data from the given file path and return it in a usable format
    loaded_data = {}
    for  key, path in path_dict.items():
        print(f"Loading data for {key} from {path}")
        # Implement actual data loading here
        loaded_data[key] = pd.read_csv(data_path+path+'/'+fname, sep=';')  # Replace with actual loaded data
    return loaded_data
```

```python
MP0_paths = {
    'Be$^{2+}$(aq)':'/MACE-MP-0/112_water/Be2+/travis',
    'BeCl$_2$(aq)':'/MACE-MP-0/112_water/BeCl2/travis',
}

MP0_data = load_data(MP0_paths, fname='rdf_Be_O_[Be1r_O1o].csv')
```

```python
MP0_data
```

```python
plt.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(figsize=(8, 6))
for key, data in MP0_data.items():
    ax.plot(data['# Distance / pm']/100, data['  g(r)'], label=key)
ax.set_title(f'RDF for Be-O (MACE-MP-0)')
ax.set_xlabel('r (Å)')
ax.set_ylabel('g(r)')
ax.legend()
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
for key, data in MP0_data.items():
    ax.plot(data['# Distance / pm']/100, data['  Integral'], label=key)
ax.set_title(f'Neighbour distribution for Be-O  (MACE-MP-0)')
ax.set_xlabel('r (Å)')
ax.set_ylabel('N(r)')
#ax.set_ylim(0, 20)
ax.grid()
ax.legend()
plt.show()
```
