```python
data_path = '../data'
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_data(path_dict, fname):
    # Placeholder for data loading logic
    # This function should read the data from the given file path and return it in a usable format
    loaded_data = {}
    for  key, path in path_dict.items():
        print(f"Loading data for {key} from {path}")
        # Implement actual data loading here
        loaded_data[key] = pd.read_csv(data_path+path+'/'+fname, sep=';')  # Replace with actual loaded data
        print(loaded_data[key].columns)  # Print the first few rows to verify loading
    return loaded_data
```

```python
MP0_paths = {
    'Be2+':'/MACE-MP-0/112_water/Be2+/travis',
    'BeCl2':'/MACE-MP-0/112_water/BeCl2/travis',
}

MP0_data = load_data(MP0_paths, fname='rdf_BeH8O4_H2O_[Ber_Oo].csv')
```

```python
MP0_data
```

```python
plt.figure()
for key, data in MP0_data.items():
    plt.plot(data['# Distance / pm']/100, data['  g(r)'], label=key)
plt.title(f'RDF for Be-O (MACE-MP-0)')
plt.xlabel('r (Å)')
plt.ylabel('g(r)')
plt.legend()
plt.show()
```

```python
plt.figure()
for key, data in MP0_data.items():
    plt.plot(data['# Distance / pm']/100, data['  Integral'], label=key)
plt.title(f'Neighbour distribution for Be-O  (MACE-MP-0)')
plt.xlabel('r (Å)')
plt.ylabel('g(r)')
plt.grid()
plt.legend()
plt.show()
```
